## Seaborn
- 통계 정보 시각화 라이브러리
- 간결한 코드로 그래프를 예쁘게 그릴 수 있음

**반면 Matplotlib은 조금 올드하지만, seaborn의 조상격**

```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/bike.csv') #자전거 사용 데이터 프레임
df
```

```python
df.describe(include='all')
```

```python
# 그래프 크기 변환 함수
# 
# plt.fiqure(figsize=(12, 6)) 는 아래와 같음
# sns.set_theme(rc={'figure.figsize' : (12, 6)})
```

```python
plt.rc('font', family='AppleGothic')
plt.figure(figsize=(12, 6))
plt.bar(df['month'], df['total'])
```

```python


sns.barplot(x=df['month'], y=df['total'])
```

```python
sns.barplot(data=df, x='month', y='total') #data=df를 앞에 붙여주면 뒤에 df머시기 다안써도됨
```

```python
#신뢰구간 : confidence  interval. ci가 그 약어
# ci = none을 넣어주면 막대가사라짐

sns.barplot(data=df, x='month', y='total', errorbar = None)
```

```python
sns.set_theme(style = "white")
```

```python
sns.barplot(data=df, x='month', y='total', errorbar = None)
```

## 월별 일하는 여부에 따른 총 대여 건수 막대 차트

```python
sns.set_theme(rc={'figure.figsize' : (12, 6)})
sns.barplot(data=df, x='month', y='total', errorbar = None, hue='workingday')
plt.show()
```

```python
sns.set_theme(rc={'figure.figsize' : (12, 6)})
sns.barplot(data=df, x='month', y='registered', errorbar = None, hue='workingday')
plt.show()
```

```python
sns.set_theme(rc={'figure.figsize' : (12, 6)})
sns.barplot(data=df, x='month', y='casual', errorbar = None, hue='workingday')
plt.show()
```

```python
sns.set_theme(rc={'figure.figsize' : (12, 6)})
sns.lineplot(data=df, x='month', y='casual', errorbar = None, hue='workingday')
plt.show()
```

```python
# 선택지가 크게는 화이트, 화이트그리드, 다크그리드 3가지 배경 옵션이 있음
```

```python
sns.set_theme()  # style='darkgrid'라는 값이 기본 값.
sns.barplot(data=df, x='month', y='total', errorbar = None, palette="rainbow")
plt.show()
```

```python
sns.set_theme(style='ticks')  # 눈금
sns.barplot(data=df, x='month', errorbar=None, y='total', palette='pink')
plt.show()
```
