```python
import pandas as pd
import matplotlib as plt
```

```python
df = pd.read_csv('../data/revenue.csv')
df
```

```python
df.sum()
```

```python
#언제쯤 빠르게 성장했는지
df['revenue'].cumsum()
```

```python
df['revenue_cumsum'] = df['revenue'].cumsum() #cubjective sum 누적합
df
```

```python
df.plot(x='month', y='revenue_cumsum')
```

```python
df.plot(x='month', y='revenue_cumsum', kind ='bar')
```
